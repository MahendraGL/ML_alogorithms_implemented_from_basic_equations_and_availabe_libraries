Plotting the Logistic or Sigmoid Function:


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Z=np.linspace(-6,6,1000)
G=1/(1+np.exp(-Z)) # G(Z)=1/1+exp(Z) is the sigmoid function here....
```

```python
plt.plot(Z,G,'r',label="Sigmoid Function Plot")
# mesh_pos=np.linspace(-5,5,10)
# plt.xticks(mesh_pos)
plt.axhline(y=0.5, color='b', linestyle='--', label="y=0.5")
plt.axvline(x=0, color='g', linestyle='--', label="x=0")
plt.grid(True, linestyle='--', alpha=0.5)
plt.grid(True, linestyle='--', alpha=0.5)
plt.xlabel("Z")
plt.ylabel("G(Z)")
plt.legend()
```

```python
dG=G* (1-G)
mean=np.mean(Z)
plt.plot(Z,dG,'r',label="Differntial of Sigmoid Function Plot")
plt.axvline(x=mean,color='g',linestyle='--', label='mean')
plt.grid(True, linestyle='--', alpha=0.5)
plt.xlabel("Z")
plt.ylabel("dG(Z)")
plt.legend()
```

Logistics Regression: 0 OR 1!!!


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv("/content/heart (1).csv")
df.head()
```

```python
X=df.age
Y=df.output
plt.scatter(X,Y,color='r')
```

```python
X_NP=X.values
X_val=X_NP.reshape(-1,1)

Y_NP=Y.values
Y_val=Y_NP.reshape(-1,1)
```

```python
LR=LinearRegression()  # What will be the result if LR is used to predict?
LR.fit(X_val,Y_val)
```

```python
X_pred=np.linspace(30,80,200).reshape(-1,1)
Y_pred=LR.predict(X_pred)
```

```python
plt.scatter(X,Y,color='r',label='Binary Classifier')
plt.plot(X_pred,Y_pred,color='b',label="Linear Reg. Line")
```

```python
Z=np.linspace(-80,80,1000)
G=1/(1+np.exp(-Z)) # G(Z)=1/1+exp(Z) is the sigmoid function here....
```

```python
plt.plot(Z,G,'r',label="Sigmoid Function Plot")
plt.scatter(X,Y,color='magenta')
# mesh_pos=np.linspace(-5,5,10)
# plt.xticks(mesh_pos)
plt.axhline(y=0.5, color='b', linestyle='--', label="y=0.5")
plt.axvline(x=0, color='g', linestyle='--', label="x=0")
plt.grid(True, linestyle='--', alpha=0.5)
plt.grid(True, linestyle='--', alpha=0.5)
plt.xlabel("Z")
plt.ylabel("G(Z)")
plt.legend()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=21)
```

```python
X_new=X_train.values
X_LR=X_new.reshape(-1,1)

Y_new=Y_train.values
Y_LR=Y_new.reshape(-1,1)

X_new_test=X_test.values
X_LR_test=X_new_test.reshape(-1,1)

Y_new_test=Y_test.values
Y_LR_test=Y_new_test.reshape(-1,1)
```

```python
Log_R=LogisticRegression(random_state=0, C=1,fit_intercept=True)
Log_R.fit(X_LR,Y_LR)
```

```python
Y_res=Log_R.predict(X_LR_test)
```

```python
Log_R.score(X_LR_test,Y_LR_test)
```

```python
#plt.scatter(X_train,Y_train,color='r',label='Training data')
plt.scatter(X_test,Y_test,label='Testing data')
plt.scatter(X_LR_test,Y_res,color='g',label='Predictions')
plt.xlabel("Age")
plt.ylabel("Probabilty of having disease")
plt.legend()
```

```python
x=np.linspace(30,80,10).reshape(-1,1)
y=Log_R.predict(x)
Log_R.score(x,y)
```
