### *Underfitting*


```python
# Import all needed libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Generate some random X,Y values
X=4*np.random.rand(100,1)-2
Y=2 + 5*X + 3*(X**2)+4*(X**3) +15* np.random.randn(100,1) # Generating values of multiple order for polynominal regression.
```

```python
poly=PolynomialFeatures(degree=3) # Order of curve to be fitted.
```

```python
LR=LinearRegression()
LR.fit(X,Y)  ## Since the Y is a 3rd order polynomial the data is underfitted with low accuracy. Which can be viewed by the plots.
```

```python
X_values=np.linspace(-2,2,100).reshape(-1,1)
Y_values=LR.predict(X_values)
```

```python
plt.scatter(X,Y,color='r')
plt.plot(X_values,Y_values,'b--')
```

## *Bestfit*

```python
X_p=poly.fit_transform(X) ## As we have described the degree of polynomial is 3 the curve is fitted correctly to good accuracy.
LR.fit(X_p,Y)
```

```python
X_values=np.linspace(-2,2,100).reshape(-1,1)
X_poly_val=poly.transform(X_values)
Y_values=LR.predict(X_poly_val)
```

```python
plt.scatter(X,Y,color='r')
plt.plot(X_values,Y_values,'b--')
```

# *Overfitting*

```python
# Generate some random X,Y values
X=4*np.random.rand(100,1)-2
Y=2 + 5*X + 3*(X**2)+4*(X**3) +15* np.random.randn(100,1) # Generating values of multiple order for polynominal regression.
```

```python
poly=PolynomialFeatures(degree=10)
```

```python
X_p=poly.fit_transform(X) ## Now let us overfit the curve with degree=5 in the polynomial function and visualize.
LR.fit(X_p,Y)
```

```python
X_values=np.linspace(-2,2,100).reshape(-1,1)
X_poly_val=poly.transform(X_values)
Y_values=LR.predict(X_poly_val)
```

```python
plt.scatter(X,Y,color='r')
plt.plot(X_values,Y_values,'b--')
```
