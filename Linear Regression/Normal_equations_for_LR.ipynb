```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=5*np.random.rand(100,1)
Y=7+3*X + np.random.randn(100,1)
```

```python
plt.plot(X,Y,'bo',color='g')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
```

```python
X_new=np.c_[np.ones((100,1)),X]
```

```python
X_new.shape
```

*Normal Equations*

```python
param=np.linalg.inv(X_new.T.dot(X_new)).dot(X_new.T).dot(Y)
```

```python
param
```

```python
Y_pre=X_new.dot(param)

X_reduced=X_new[:, 1:2]
```

```python
plt.plot(X_reduced,Y_pre,'g--',label="Regression Line using Normal Eq.")
plt.plot(X,Y,'bo')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend()
```

```python
#Predictions
x=np.array([1.2,3.5,4,2.2])

for i in (x):
  y=param[0]+param[1]*i
  print(y)
```
